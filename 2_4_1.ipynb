```
import numpy as np
from matplotlib import pyplot as plt
```

```
# Método de Runge Kutta

def rk4( func, h, ti, tf, x0, par ):
    ti = ti
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tf ):

      x.append (xi)
      t.append (ti)
      k1 = func( ti,xi,par)
      k2 = func( ti + 0.5*h,xi + 0.5*k1*h,par )
      k3 = func( ti + 0.5*h,xi + 0.5*k2*h,par)
      k4 = func( ti + h,xi + k3*h,par) 

      xi = xi + h*(k1 + 2*k2 + 2*k3 + k4)/6
      ti = ti + h
    return t, x


# Método de Euler 

def euler( func, h, ti, tf, x0, par ):
    ti = ti 
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tf ):
    
      x.append (xi)
      t.append (ti)
  
      xi = xi + h*func(ti,xi,par)
      ti = ti + h
    return t, x
```
